# Kauai Amakihi - Song Visual Analysis
below we have the distinct song of the Kauai Amakihi

```python
# %%

from opensoundscape import Audio,audio,Spectrogram
path = "/home/dhind/Kauai-Amakihi/kauai-amakihi-song-isolated.wav"
sampler_audio = Audio.from_file(path)
sampler_audio.show_widget(normalize=True)
spectrogram_object = Spectrogram.from_audio(sampler_audio)
spectrogram_object.plot()
```

Below I have demonstrated 55 window samples, 1100 window samples, 2048 window samples.

```python
audio = sampler_audio.resample(22000)
spec_55 = Spectrogram.from_audio(audio, window_samples=55, overlap_samples=0)
spec_55.plot()
#sample of 55 window samples
spec_1100 = Spectrogram.from_audio(audio, window_samples=1100, overlap_samples=0)
spec_1100.plot()
#sample of 1100 window sample
spec_2048 = Spectrogram.from_audio(audio, window_samples=2048, overlap_samples=0)
spec_2048.plot()
#sample of 11
```
